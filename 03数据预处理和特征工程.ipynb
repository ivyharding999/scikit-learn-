## 2.1数据无量纲化

```python
from sklearn.preprocessing import MinMaxScaler
data = [[-1,2],[-0.5,6],[0,10],[1,18]]
```

```python
import pandas as pd
pd.DataFrame(data)
```

```python
# 实现归一化
scaler = MinMaxScaler()  #实例化
scaler = scaler.fit(data) #fit,在这里本质上是生成min(x)和max(x)
result = scaler.transform(data)#通过接口导出结果，计算公式
result
```

```python
# scaler.fit_transform(data)可以实现上面两句代码的结果
result_ = scaler.fit_transform(data)#训练和导出结果一步达成
result_
```

```python
data_ = scaler.inverse_transform(result)#将归一化后的结果逆转
data_
```

也可以使用MinMaxScaler的参数feature_range实现将数据归一化到[0，1]以外的范围内

```python
data = [[-1,2],[-0.5,6],[0,10],[1,18]]
scaler = MinMaxScaler(feature_range=[5,10])#依然实例化
result2 = scaler.fit_transform(data)
result2
```

当x中的特征数量非常多的时候，fit会报错表示，数据量太大了计算不了了。此时使用partial_fit作为训练接口
scaler = scaler.partial_fit(data)

## BONUS(添加):使用numpy来实现归一化

```python
import numpy as np
x = np.array([[-1,2],[-0.5,6],[0,10],[1,18]])
# 归一化
x_nor = (x-x.min(axis=0))/(x.max(axis=0)-x.min(axis=0))
x_nor
```

```python
# 逆转归一化
x_returned = x_nor*(x.max(axis=0)-x.min(axis=0))+x.min(axis=0)
x_returned
```

# 3.特征选择

```python
import pandas as pd 
data = pd.read_csv("E:/ivy_machine_learning/program/莫烦学习/2019年寒假学习+sklearn+jupyter用法/菜菜sklearn完整版/digit recognizor.csv")
data.head()
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
print(X.shape,y.shape)
```

## 3.1 filter 过滤法
### 3.1.1 方差过滤法

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
X_var0 = VarianceThreshold().fit_transform(X)
X_var0.shape
```

默认阈值是0，过滤了近80个特征（说明这些特征不含任何信息）

如果我们想要过滤掉一半的特征，那我们可以将阈值设置为方差的中位数

```python
import numpy as np
x_fvar = VarianceThreshold(np.median(X.var().values)).fit_transform(X)
x_fvar.shape
```

#### 方差过滤对模型的影响

KNN VS RandomForest 在不同方差过滤下效果的对比

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
cross_val_score(RandomForestClassifier(n_estimators=10,random_state=0),X,y,cv=5).mean()
```

```python
cross_val_score(RandomForestClassifier(n_estimators=10,random_state=0),x_fvar,y,cv=5).mean()
```

KNN方差过滤前后对比

```python
%%timeit
knn_score1 = cross_val_score(KNeighborsClassifier(),X,y,cv=5).mean()
print(knn_score1)
```

### 3.1.2 相关性过滤
判断特征和标签之间的相关性：卡方，F检验，互信息

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
# 假设我们需要300个特征
x_fschi = SelectKBest(chi2,k=300).fit_transform(X,y)
x_fschi.shape
```

```python
cross_val_score(RandomForestClassifier(n_estimators=10,random_state=0),x_fschi,y,cv=5).mean()
```

这个相比之前的392个特征，结果变低了，说明过滤掉了一些有效特征，所以这个参数K的选择比较重要。可以通过学习曲线去找合适的k
