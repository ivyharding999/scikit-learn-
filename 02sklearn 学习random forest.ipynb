## sklearn 建模的基本流程

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(wine.data,wine.target,test_size=0.3)
clf =DecisionTreeClassifier(random_state=0)  #实例化，random_state=0为了固定生成一棵树
rfc = RandomForestClassifier(random_state=0)  #实例化，random_state=0为了固定生成一片森林

clf = clf.fit(x_train,y_train)
rfc = rfc.fit(x_train,y_train)

score_c = clf.score(x_test,y_test)
score_r = rfc.score(x_test,y_test)
#将字符串和数字整体输出：用format实现---{}里面填入score_c的内容
print("single tree:{}".format(score_c),"randomforest:{}".format(score_r))
```

## 画出随机森林和决策树在十组交叉验证下的效果对比

### 交叉验证：将数据集换分为n分，依此取每一份做测试机，每n-1分做训练集，多次训练模型以及观测模型稳定性的方法。

```python
from sklearn.model_selection import cross_val_score
rfc_l = []
clf_l = []
for i in range(10):
    rfc = RandomForestClassifier(n_estimators=25)
    rfc_s = cross_val_score(rfc,wine.data,wine.target,cv=10).mean()
    rfc_l.append(rfc_s)
    clf = DecisionTreeClassifier()
    clf_s = cross_val_score(clf,wine.data,wine.target,cv=10).mean()
    clf_l.append(clf_s)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),rfc_l,label="Random Forest")
plt.plot(range(1,11),clf_l,label="Decision Tree")
plt.legend()
plt.show()
```

### 观察上图，我们发现，单个决策树的波动轨迹和随机森林一致。
### 再次验证了单个决策树的准确率越高，随机森林的准确率也越高。

## n_estimators的学习曲线

```python
superpa = []
for i in range(200):
    rfc = RandomForestClassifier(n_estimators=i+1,n_jobs=-1)
    rfc_s = cross_val_score(rfc,wine.data,wine.target,cv=10).mean()
    superpa.append(rfc_s)
print(max(superpa),superpa.index(max(superpa)))
plt.figure(figsize=[20,5])
plt.plot(range(1,201),superpa)
plt.show()
```

# 实例：随机森林在乳腺癌数据上的调参

## 1 需要导入的库

```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 2 导入数据

```python
data = load_breast_cancer()

data
```

```python
data.data.shape
```

```python
data.target.shape
```

## 3 第一次见简单建模

```python
rfc = RandomForestClassifier(n_estimators=100
                             ,random_state=30
                            )
score_pre = cross_val_score(rfc,data.data,data.target,cv=10).mean()

score_pre
```

## 4 随机森林调整的第一步：无论如何先来调n_estimators

```python
scorel = []
for i in range(0,200,10):
    rfc = RandomForestClassifier(n_estimators=i+1,
                                n_jobs=-1,
                                random_state=30
                                )
    score = cross_val_score(rfc,data.data,data.target,cv=10).mean()
    scorel.append(score)
print(max(scorel),(scorel.index(max(scorel)))*10+1)
plt.figure(figsize=[20,5])
plt.plot(range(1,201,10),scorel)
plt.show()
###list.index([object])
###返回object在列表list中的索引
```

## 5 在确定好的范围内，进一步细化学习曲线

```python
scorel = []
for i in range(125,140):
    rfc = RandomForestClassifier(n_estimators=i,
                                n_jobs=-1,
                                random_state=30
                                )
    score = cross_val_score(rfc,data.data,data.target,cv=10).mean()
    scorel.append(score)

plt.figure(figsize=[20,5])
plt.plot(range(125,140),scorel)
plt.show()
###list.index([object])
###返回object在列表list中的索引
```

```python
print(max(scorel),([*range(125,140)][scorel.index(max(scorel))]))
```

## 6 为网格搜索做准备，书写网格搜索的参数

```python
"""
有些参数是没有参照的，很难说清楚一个范围，这种情况下我们使用学习曲线，看趋势从曲线跑出的
结果中选取一个更小的区间，再跑曲线。
param_grid = {'n_estimators':np.arange(0,200,10)}
param_grid = {'max_depth':np.arange(1,20,1)}
param_grid = {'max_leaf_nodes':np.arange(25,50,1)}
    对于大型数据集，可以尝试从1000来构建，每100个叶子节点一个区间，再
逐渐所辖范围。 
有些参数可以找到一个范围，或者说是我们知道他们的取值和随着他们的取值，模型的征途准确率
会如何变化，这样的参数我们可以直接跑网格搜索
param_grid = {'criterion':['gini','entropy']}
param_grid = {'min_samples_split':np.arange(2,2+20,1)}
param_grid = {'min_samples_leaf':np.arange(1,1+10,1)}

"""
```

我们试下用dict来一次性确定参数

```python
param_grid = dict(max_depth=np.arange(1,20,1)
                 ,max_leaf_nodes=np.arange(25,50,1)
                 ,criterion=['gini','entropy']
                 ,min_samples_split = np.arange(2,2+20,1)
                 ,min_samples_leaf = np.arange(1,1+10,1)
                 )
rfc = RandomForestClassifier(n_estimators=129
                            ,random_state=30
                            )
GS = GridSearchCV(rfc,param_grid,cv=10)
GS.fit(data.data,data.target)
```

```python
print("The best parameters are %s with a score of %f"%(GS.best_params_,GS.best_score_))
```

## 7 开始按照参数对模型整体准确率的影响程度进行调参，首先调整max_depth

```python
# 调整max_depth
param_grid = {'max_depth':np.arange(1,20,1)}
"""
一般根据数据的大小来进行一个试探，乳腺癌数据很小，所以可以采用1~10，或者
可以采用1~20这样的试探。但是对于digit recognition这样的大型数据来说，我们应该
尝试30~50层深度（或许还不足够，更应该画出学习曲线，来观察深度对模型的影响
"""
rfc = RandomForestClassifier(n_estimators=129
                            ,random_state=30
                            )
GS = GridSearchCV(rfc,param_grid,cv=10)
GS.fit(data.data,data.target)
```

```python
GS.best_params_  # 发现通过max_depth这个参数想让正确率增加，但是结果却降低了，回想泛化误差曲线，我们可以得到：
                 # 我们想让这个模型向左移动，但是正确率降低，说明这个模型本身就是够简单了，不能通过更改max_depth
```

```python
GS.best_score_
```

## 8调整max_features

```python
#调整max_features
param_frid = {'max_features':np.arange(5,30,1)}
"""
max_features是唯一能够将模型往左（低方差高偏差）推，也能
够将模型往右推的参数。我们需要根据调参前，模型所在的位置（
在泛化误差最低点的而左边还是右边）来决定我们要将max_features
往哪边调。现在模型位于图像的左侧，我们需要的是更高的复杂度，因此
我们应该把max_features往更大的方向调整，可用的特征越多越复杂。
max_features的默认最小值是sqrt(n_features)，因此我们使用这个值
作为调参范围的最小值。
"""
rfc = RandomForestClassifier(n_estimators=129
                            ,random_state=30
                            )
GS = GridSearchCV(rfc,param_grid,cv=10)
GS.fit(data.data,data.target)
```

```python
GS.best_params_
```

```python
GS.best_score_
```

## 9调整min_samples_leaf

```python
# 调整min_samples_leaf
param_grid = {'min_samples_leaf':np.arange(1,1+10,1)}
"""
对于min_samples_split和min_samples_leaf，一般都是从他们的
最小值开始向上增加10和20，面对高纬度样本量数据，如果不放心，也可以
直接+50，对于大型数据，可能需要200~300的范围。如果调整的时候发现准确率
无论如何都上不来，那么可以放心大胆的调一个很大的数据，大力限制模型的复杂度
"""
rfc = RandomForestClassifier(n_estimators=129
                            ,random_state=30
                            )
GS = GridSearchCV(rfc,param_grid,cv=10)
GS.fit(data.data,data.target)
```

```python
GS.best_params_
```

```python
GS.best_score_
```

## 10 继续尝试min_samples_split

```python
#调整min_samples_split
param_grid = {'min_samples_split':np.arange(2,2+20,1)}
rfc = RandomForestClassifier(n_estimators=129
                            ,random_state=30
                            )
GS = GridSearchCV(rfc,param_grid,cv=10)
GS.fit(data.data,data.target)
```

```python
GS.best_params_
```

```python
GS.best_score_
```

## 11 最后尝试criterion

```python
#调整Criterion
param_grid = {'criterion':['gini','entropy']}
rfc = RandomForestClassifier(n_estimators=129
                            ,random_state=30
                            )
GS = GridSearchCV(rfc,param_grid,cv=10)
GS.fit(data.data,data.target)
```

```python
GS.best_params_
```

```python
GS.best_score_
```

```python
rfc = RandomForestClassifier(n_estimators=129
                             ,criterion = 'entropy'
                             ,random_state=30
                            )
score = cross_val_score(rfc,data.data,data.target,cv=10).mean()
score
```

# 3.2 实例，用随机森林回归填补缺失值

```python
data = load_boston()
print(data.data.shape,data.target.shape)
```

```python
x_full, y_full = data.data,data.target
n_samples = x_full.shape[0]
n_features = x_full.shape[1]
```

为完整数据假如缺失值，假设比例是50%，需要有3289.0个数据缺失

```python
rng = np.random.RandomState(0)
missing_rate = 0.5
n_missing_samples = int(np.floor(n_samples*n_features*missing_rate))
```

```python
missing_features = rng.randint(0,n_features,n_missing_samples)
missing_samples = rng.randint(0,n_samples,n_missing_samples)
```

```python
x_missing = x_full.copy()
y_missing = y_full.copy()

x_missing[missing_samples,missing_features] = np.nan

x_missing = pd.DataFrame(x_missing)
```
