```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

## 读数据

```python
wine = load_wine()

# feature:wine.data
# target:wine.target
```

```python
import pandas as pd
pd.concat([pd.DataFrame(wine.data),pd.DataFrame(wine.target)],axis=1)
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
x_train,x_test,y_train,y_test = train_test_split(wine.data,wine.target,test_size=0.3)
```

```python
print("x_train:",x_train.shape)
print("x_test:",x_test.shape)
print("y_train:",y_train.shape)
print("y_test:",y_test.shape)
```

## 建模

```python
clf = tree.DecisionTreeClassifier(criterion="entropy")  #实例化，criterion默认是gini系数
clf = clf.fit(x_train,y_train)   # 训练模型
score = clf.score(x_test,y_test)   # 使用接口.score来获取分类的正确率 ，这里我觉得是     预测正确的/总的
score
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons,make_circles,make_classification
from sklearn.tree import DecisionTreeClassifier
```

### 生成三种数据集
##### 我们先从sklearn自带数据库中生成三种类型的数据集：（1）月亮型数据，（2）环形数据，（3）二分型数据

```python
# make_classification库生成随机的二分型数据
x,y = make_classification(n_samples=100,#生成100个样本
                          n_features=2,#包含2个特征，即生成二维数据
                          n_redundant=0,#添加冗余特征0个
                          n_informative=2,#包含信息的特征是0个
                          random_state=1,#随机模式1
                          n_clusters_per_class=1#每个簇包含的标签类别有1个
                         )
#在这里可以观察下x,y，其中x是100行带有2个特征的数据，y是二分类标签
#也可以画出散点图来观察下x中的特征的分布
```

```python
print("x:",x.shape)
print("y:",y.shape)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
#从图上看，生成的二分数据的2个簇离得彼此很远，这压根不利于我们测试分类器的效果，
#因此，我们使用np生成随机数组，通过已经生成的二分型数据加减0—1之间的随机数，使数
#据分布变得很稀疏
#注意，这个过程只能进行一次，因为多次运行之后的x会变得非常稀疏，2个簇的数据会混合
#在一起，分类器的效果会持续下降。
```

```python
rng = np.random.RandomState(2)#生成一种随机模式
x += 2*rng.uniform(size=x.shape)#加减0-1之间的随机数
linearly_separable = (x,y)#生成新的x,依然可以画散点图来观察特征
plt.scatter(x[:,0],x[:,1])
```

```python
#用make_moons创建月亮型数据，make_circles创建环形数据，
#并将三组数据打包起来放在列表datasets中
datasets = [make_moons(noise=0.3,random_state=0),
           make_circles(noise=0.2,factor=0.5,random_state=1),
           linearly_separable]
```
