# 3. 处理样本数据极度不平衡
类别不平衡问题，顾名思义，即数据集中存在某一类样本，其数量远多于或远少于其他类样本，从而导致一些机器学习模型失效的问题。例如逻辑回归即不适合处理类别不平衡问题，例如逻辑回归在欺诈检测问题中，因为绝大多数样本都为正常样本，欺诈样本很少，逻辑回归算法会倾向于把大多数样本判定为正常样本，这样能达到很高的准确率，但是达不到很高的召回率。

类别不平衡问题在很多场景中存在，例如欺诈检测，风控识别，在这些样本中，黑样本（一般为存在问题的样本）的数量一般远少于白样本（正常样本）。

原文链接：https://blog.csdn.net/nlpuser/article/details/81265614解决方案：过采样或者下采样

下采样：将标签为0的样本数量下降到和标签为1样本数量相等，也就是样本同样少（500左右）

过采样：将标签为1的样本数量上升到和标签为0样本数量相等，也就是样本同样多（28万左右）

# 5. 建模
采用逻辑回归

评估标准采用：召回率recall，或者精度precision

假如正例：990 负例：10  

假如经过一个模型后，我们发现990个正例我们都预测对了，而负例都错了

此时：精度precision = 预测正确的/总数=990/1000=99%，虽然很高，但是我们的任务是找到负例，显然不服从我们的需求

对于召回率recall = 0/10 = 0，我们的召回率为0；

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold,cross_val_score
from sklearn.metrics import confusion_matrix,recall_score,classification_report
```

### 调参，获得最优的C

```python
def printing_Kfold_scores(X_train_data,y_train_data):
    fold = KFold(n_splits=5)
    #设置不同的参数C
    c_parameter_range = [0.01,0.1,1,10,100]
    
    result_table = pd.DataFrame(index=range(len(c_parameter_range),2),columns=["c_parameter","mean recall score"])
    result_table['c_parameter'] = c_parameter_range
    
    j = 0
    for c_parameter in c_parameter_range:
        print("---"*10)
        print("c_parameter=",c_parameter)
        print("---"*10)
        print('')
        
        recall_accs = []
        for iteration, indices in enumerate(fold.split(X_train_data)):
            lr = LogisticRegression(C=c_parameter,penalty='l1')#实例化
            # 训练
#             y_temp = y_train_data.iloc[indices[0],:].values.ravel()
            print(X_train_data.shape)
            print(y_train_data.shape)
            X_temp = X_train_data.iloc[indices[0],:]
            print("X_temp.shape",X_temp.shape)
            y_temp = y_train_data.iloc[indices[0]]
            print("y_temp.shape",y_temp.shape)
            lr.fit(X_temp,y_temp)
            # 预测
            y_pred_undersample = lr.predict(X_train_data.iloc[indices[1],:].values)
            # 从召回率计算
            recall_acc = recall_score(y_train_data.iloc[indices[1]].values,y_pred_undersample)
            recall_accs.append(recall_acc)
            print("iteration=",iteration,"recall_acc=",recall_acc)
        
        result_table.ix[j,"mean recall score"] = np.mean(recall_accs)
        j += 1
        print('')
        print("mean recall score",np.mean(recall_accs))
        print('')
    print(result_table)
    s1 = result_table["mean recall score"]
    max_c_index = (s1[s1==s1.max()].index).values
    print("max_c=",max_c_index)
    
    best_c = result_table.iloc[max_c_index,0].values
    print("***"*10)
    print("best c_parameter=",best_c)
    print("***"*10)
    return best_c
# 方式2：直接用cross_val_score，舍弃KFold
def printing_cross_val_scores(X_train_data,y_train_data):
    #设置不同的参数C
    c_parameter_range = [0.01,0.1,1,10,100]
    
    result_table = pd.DataFrame(index=range(len(c_parameter_range),2),columns=["c_parameter","mean recall score"])
    result_table['c_parameter'] = c_parameter_range
    
    j = 0
   
    for c_parameter in c_parameter_range:
        print("---"*10)
        print("c_parameter=",c_parameter)
        print("---"*10)
        print('')
        recall_accs = []
        for time in range(5):
            lr = LogisticRegression(C=c_parameter,penalty='l1')#实例化
            # 训练
            recall_acc = cross_val_score(lr,X_train_data,y_train_data,cv=5,scoring="recall").mean()
        print("recall_acc=",recall_acc)
        recall_accs.append(recall_acc)

        result_table.ix[j,"mean recall score"] = recall_acc
        j += 1
        print('')
        print("mean recall score",recall_acc)
        print('')
    print(result_table)
    s1 = result_table["mean recall score"]
    max_c_index = (s1[s1==s1.max()].index).values
    print("max_c=",max_c_index)
    
    best_c = result_table.iloc[max_c_index,0].values
    print("***"*10)
    print("best c_parameter=",best_c)
    print("***"*10)
    return best_c
```

### 采用混淆矩阵

```python
def plot_confusion_matrix(cm, title):
#     cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]    # 归一化   
#     plt.figure(figsize=(10,10))
    plt.imshow(cm
#                , interpolation='nearest'
              )    # 在特定的窗口上显示图像    
    plt.title(title)    # 图像标题    
    plt.colorbar()    
    plt.ylabel('True label')# 坐标轴标签
    plt.xlabel('Predicted label')# 坐标轴标签
    plt.text(0, 0.1,cm[0][0])
    plt.text(0.9, 0.1,cm[0][1])
    plt.text(0, 1.1,cm[1][0])
    plt.text(0.9, 1.1,cm[1][1])
   
```

```python
# 在下采样数据集上测试
lr = LogisticRegression(C=0.01,penalty='l1')#实例化
lr.fit(X_downsample_train,y_downsample_train)
# 预测
y_pred_undersample = lr.predict(X_downsample_test)
#混淆矩阵，正确结果是在对角线上的，对角线外的元素越少越好
cm = confusion_matrix(y_downsample_test,y_pred_undersample)
cm
```
